## Circular Kalman filter with different weights

```python
import kalman.filtering as flt

n_point = int((T/dt) + 1)

# Parameters
# kappa_y_set = np.array([10, 10, 10, 10, 10, 10, 10, 10])
# alpha_set = np.array([10, 1, 0.08, 0.058, 0.057, 0.05, 0.01, 0])
kappa_y_set = 5*np.array([10, 10, 10, 10, 10, 10, 10, 10])
alpha_set = np.array([10, 1, 0.4, 0.27, 0.25, 0.2, 0.1, 0])

kappa_phi = 1 # inverse diffusion constant
phi_0 = z[0] # initial mean
kappa_0 = 20 # initial certainty

# Data
z = utility.wrapToPi(df.gamma)
z_filt = utility.wrapToPi(df.gamma_filt)
dy = df.dphi * dt

dy = np.zeros(n_point)
zz = np.unwrap(df.psi)
for i in range(1,n_point):
    dy[i] = np.random.normal(zz[i]-zz[i-1], 0*1/np.sqrt(kappa_y) * np.sqrt(dt)) 

# Loop
n_set = len(kappa_y_set)
gamma_kalman = np.zeros((n_point, n_set))
for n in range(0,n_set):
    # Parameters
    kappa_y = kappa_y_set[n] # precision of increment observations
    alpha = alpha_set[n] # precision of direct observations, lin apprx
    
    # Run filter
    mu_VM, kappa_VM = flt.vM_Projection_Run(T,
                                            kappa_phi,
                                            z=z,
                                            alpha=alpha,
                                            dy=dy,
                                            kappa_y=kappa_y,
                                            phi_0=phi_0,
                                            kappa_0=kappa_0,
                                            dt=dt)
    
    gamma_kalman[:,n]= np.squeeze( utility.wrapToPi(mu_VM) )
    
# Figure
mksz = 0.5
linestyle = ''
marker = '.'
fig, ax = plt.subplots(n_set, 1, figsize=(2,n_set*0.85), dpi=300)
for n in range(0,n_set):
    ax[n].plot(*flt.circplot(df.t, df.gamma.values), label='Raw $\gamma$', color='gray',
                 linestyle='',
                 linewidth=1,
                 marker='.',
                 markersize=mksz)

    ax[n].plot(*flt.circplot(df.t, df.gamma_filt.values), label='Butterworth filter $\gamma$', color='black',
                 linestyle=linestyle,
                 linewidth=1.5,
                 marker=marker,
                 markersize=mksz)

    ax[n].plot(*flt.circplot(df.t, df.psi.values), label='$\psi$', color='blue',
                 linestyle=linestyle,
                 linewidth=1.5,
                 marker=marker,
                 markersize=mksz)

    ax[n].plot(*flt.circplot(df.t, gamma_kalman[:,n]), label='Circular kalman filter $\gamma$', color='orange',
                 linestyle=linestyle,
                 linewidth=1.5,
                 marker=marker,
                 markersize=mksz)

    ax[n].tick_params(axis='x', labelsize=6)
    ax[n].tick_params(axis='y', labelsize=6)
    ax[n].set_ylabel('$(\circ)$', fontsize=6)
    ax[n].grid(color='gray', linestyle='-', linewidth=0.75, alpha=0.2)

    ytickpispace = 0.5
    ax[n].set_ylim(-1*np.pi, 1*np.pi)
    ylim = ax[n].get_ylim()
    yticks = np.arange(ylim[0], ylim[1]+0.01, ytickpispace*np.pi)
    ytickpi = yticks / np.pi
    ytickslabels = ytickpi.tolist()
    for y in range(len(ytickslabels)):
        ytickslabels[y] = ('$'+ str(Fraction(ytickslabels[y])) + '\pi$')

    ax[n].set_yticks(yticks)
    ax[n].set_yticklabels(ytickslabels)
    
    ax[n].set_title(r'$\alpha$ = ' + str(alpha_set[n]) + ', ' + r'$\kappa$ = ' + str(kappa_y_set[n]), 
                    fontdict={'fontsize': 6, 'fontweight': 'medium'}, y=1.5, pad=-14)
    # ax[n].title.set_text()
    
    if n == 0:
        ax[n].legend(prop={'size': 5})
        ax[n].legend(bbox_to_anchor=(1.1, 1.05))
        
    if n < n_set:
        ax[n].tick_params(axis='x', labelbottom=[])
    
    if n == n_set-1:
        ax[n].set_xlabel('time (s)', fontsize=8)
        
plt.subplots_adjust(hspace=0.5)

left  = 0.4    # the left side of the subplots of the figure
right = 1.5    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.3   # the amount of width reserved for blank space between subplots
hspace = 0.5   # the amount of height reserved for white space between subplots

plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)

plt.savefig(figname + "_circ_kalman_sweep.png", bbox_inches='tight')
plt.show()
```

## Wind direction calculations

```python
def contdirest(phi, psi, gamma, dphi, dpsi, dgamma):
    delta = gamma - psi
    ddelta = dgamma - dpsi
    
    zeta = 2*np.arctan((ddelta - np.sqrt( (ddelta**2) + ((np.tan(delta)**2) * ( (dgamma**2) + 2*dgamma*dphi + (dphi**2) )) )) 
    / (dgamma + dphi*np.tan(delta))) + phi + gamma
    
    return zeta

zeta_contdirest = contdirest(df.phi_filt, df.psi, df.gamma_filt, df.dphi, df.dpsi, df.dgamma)
zeta_contdirest = np.array(zeta_contdirest)
zeta_contdirest_unwrap = np.unwrap(zeta_contdirest)

plt.plot(time, zeta_contdirest_unwrap, '.')

b, a = signal.butter(5, 0.2 / (fs/2))
zeta_contdirest_filt = signal.filtfilt(b, a, np.squeeze(zeta_contdirest_unwrap))
zeta_contdirest_filt = np.swapaxes(np.array(zeta_contdirest_filt, ndmin=2), 0, 1)

plt.plot(time, zeta_contdirest_filt, '.')

zeta_contdirest = utility.wrapToPi(zeta_contdirest)
zeta_contdirest_filt = utility.wrapToPi(zeta_contdirest_filt)

plt.plot(time, zeta_contdirest, '.')
plt.plot(time, zeta_contdirest_filt, '.')
```
